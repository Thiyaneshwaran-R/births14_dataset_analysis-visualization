```python
#outliers detection
columns=['age', 'Medu', 'Fedu', 'traveltime', 'studytime', 'failures', 'famrel',
       'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences', 'G1', 'G2','G3']
for col in columns:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1

    LOWER_BOUND=Q1-1.5*IQR
    UPPER_BOUND=Q3+1.5*IQR
    outliers=df[(df[col]<LOWER_BOUND) | (df[col]>UPPER_BOUND)]
    print(outliers.shape[0])
```

```python
#clipping
for col in columns:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1

    LOWER_BOUND=Q1-1.5*IQR
    UPPER_BOUND=Q3+1.5*IQR
    outliers=df[(df[col]<LOWER_BOUND) | (df[col]>UPPER_BOUND)]
    print(outliers.shape[0])
    df[col]=df[col].clip(lower=LOWER_BOUND,upper=UPPER_BOUND)
```

```python
#Bar Chart
colors=['lightcoral','lightgreen']
plt.figure(figsize=(12,14 ))
avg_performance_by_internetusers = df.groupby("internet")[["G1","G2"]].mean().reset_index()
avg_performance_by_internetusers.set_index("internet").plot(kind='bar',color=colors,width=0.5)
plt.title("Average Performance by Internet Users")
plt.ylabel("Average Performance")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,14 ))
avg_performance_by_internetusers = df.groupby("internet")["G3"].mean().reset_index()
avg_performance_by_internetusers.set_index("internet").plot(kind='bar',color='lightblue',width=0.5)
plt.title("Average Performance by Internet Users")
plt.ylabel("Average Performance")
plt.tight_layout()
plt.show()
```

```python
colors=['cyan', 'magenta']
plt.figure(figsize=(12,14 ))
avg_performance_by_activities = df.groupby("activities")[["G1","G2"]].mean().reset_index()
avg_performance_by_activities.set_index("activities").plot(kind='bar',color=colors,width=0.5)
plt.title("Average Performance by Students who attending activities")
plt.ylabel("Average Performance")
plt.tight_layout()
plt.show()
```
